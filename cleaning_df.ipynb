```python
import pycountry
ec="ec"
print(pycountry.countries.get(alpha_2="EC").name)
```

```python
historic_df["region"] = historic_df["region"].str.upper()
historic_df.head()
```

```python
list_alpha_2 = [i.alpha_2 for i in list(pycountry.countries)] 
def country_flag(df):
    if (df['region'] in list_alpha_2):
        return pycountry.countries.get(alpha_2=df['region']).name
    else:
        return 'Global'

historic_df['country_name']=historic_df.apply(country_flag, axis = 1)
```
